```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# croping the images 

parrot = cv2.imread("parrot.jpg")
elephant = cv2.imread("Elephant.jpg")
```

```python
# shape and dimensions of images
print("parrot image:- ",parrot.shape)
print("Elephant image:- ",elephant.shape)
```

### cropping images 

```python
# we will crop it bring to smaller size
# 1. parrot

parrot1 = parrot[500:3500,500:2000]
plt.imshow(parrot1)
```

```python
# we will crop it bring to smaller size
# 2. elephant

elephant1 = elephant[120:,100:]
plt.imshow(elephant1)
```

### image collage 

```python
# make both photos same size
# parrot has more size so we will make parrot image similar to elephabt

parrot_cropped = parrot[0:3495,0:2621]
plt.imshow(parrot_cropped)
```

```python
## both images have same size and dimensions
print(parrot_cropped.shape,elephant.shape)
```

```python
# Now we collage two images using numpy array

collage = np.hstack((parrot_cropped,elephant))
plt.imshow(collage)
```

```python
# vertical stack
collage = np.vstack((parrot_cropped,elephant))
plt.imshow(collage)
```

### Some more Example but with normal windo 

```python
#1. horizontal 

Green_parrot = cv2.imread("Green_parrot.jpg")
Tree = cv2.imread("Tree.jpg")
Tree = Tree[0:874,0:640]

collage = np.hstack((Green_parrot,Tree))
cv2.imshow("collage",collage)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
##2.vertical

Green_parrot = cv2.imread("Green_parrot.jpg")
Tree = cv2.imread("Tree.jpg")

Tree = Tree[0:300,0:640]
Green_parrot=Green_parrot[0:470,0:640]

collage = np.vstack((Green_parrot,Tree))
cv2.imshow("collage",collage)
cv2.waitKey()
cv2.destroyAllWindows()
```
